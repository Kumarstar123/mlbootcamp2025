```python
# Fetch the JSON data from the URL
url = 'https://raw.githubusercontent.com/shashankbl/mlbootcamp2025flc/main/extras/grocery_inventory.json'  # Use 'raw.githubusercontent.com'
try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes
    data = response.json()
except requests.exceptions.RequestException as e:
    print(f"Error fetching JSON from URL: {e}")
    #data = [{"error":"404"}]  # Initialize as empty dictionary in case of error
    #data = [{}]
    data = {}
# Convert the JSON data to a Pandas DataFrame
try:
    inventory_df = pd.DataFrame(data)
    display(inventory_df) # Display the DataFrame using display() for pretty print
except (ValueError, KeyError) as e:
    print(f"Error converting JSON to DataFrame: {e}")
    if data:
        print("JSON data loaded but could not be formatted as a table")
```

```python
from datetime import datetime

TRANSACTION_FILE = "transactions.csv"

def generate_receipt(basket, inventory):
    """Generate and display the final receipt"""

    print("\n=====  FINAL RECEIPT  =====")

    total_cost = 0
    for index, row in basket.iterrows():
        item = row["item"]  # Lowercase column name
        quantity = row["quantity"]  # Lowercase column name

        # Find the price of the item in the inventory
        price_row = inventory[inventory["item"] == item]  # Lowercase column name

        if not price_row.empty:
            price = float(price_row.iloc[0]["unit_cost"])  # Lowercase column name
            total_price = price * quantity
            total_cost += total_price
            print(f"{item}: {quantity} x ${price:.2f} = ${total_price:.2f}")
        else:
            print(f"{item}: Price not found!")

    print(f"\nTotal Cost: ${total_cost:.2f}")
    print("================================")

    return total_cost

def log_transaction(basket, total_cost, inventory):
    """Log transaction details into transactions.csv"""
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    items = []
    quantities = []

    for index, row in basket.iterrows():
        items.append(row["item"])  # Lowercase column name
        quantities.append(str(row["quantity"]))  # Lowercase column name

    transaction_data = {
        "Timestamp": [timestamp],
        "Items": [", ".join(items)],
        "Quantities": [", ".join(quantities)],
        "Total Cost": [total_cost]
    }

    transaction_df = pd.DataFrame(transaction_data)

    if os.path.exists(TRANSACTION_FILE):
        transaction_df.to_csv(TRANSACTION_FILE, mode='a', header=False, index=False)
    else:
        transaction_df.to_csv(TRANSACTION_FILE, index=False)

    # print the updated inventory
    print("\n=== UPDATED INVENTORY ===")
    display(inventory)  # Use lowercase column names
```

```python
class GroceryBasket:

    def __init__(self, basket_file="basket.csv", inventory_file="https://raw.githubusercontent.com/shashankbl/mlbootcamp2025flc/main/extras/grocery_inventory.json"):
        self.basket_file = basket_file
        self.inventory_file = inventory_file
        self.load_inventory()
        self.load_basket()

    def load_inventory(self):
        """Fetch inventory from the URL and load it into a DataFrame."""
        url = 'https://raw.githubusercontent.com/shashankbl/mlbootcamp2025flc/main/extras/grocery_inventory.json'  # Use 'raw.githubusercontent.com'
        try:
            response = requests.get(url)
            response.raise_for_status()  # Raise an exception for bad status codes
            data = response.json()
        except requests.exceptions.RequestException as e:
            print(f"Error fetching JSON from URL: {e}")
            data = []  # Initialize as an empty list in case of error

        # Convert the JSON data to a Pandas DataFrame
        if data:
            try:
                self.inventory_df = pd.DataFrame(data)
                print("Inventory loaded successfully.")
                # Normalize column names to lowercase for consistency
                self.inventory_df.columns = [col.lower() for col in self.inventory_df.columns]

                print("\n=== Grocery Inventory ===")
                display(self.inventory_df)  # Display the DataFrame using display() for pretty print

            except (ValueError, KeyError) as e:
                print(f"Error converting JSON to DataFrame: {e}")
        else:
            print("No inventory data found or error occurred. Inventory is empty.")
            self.inventory_df = pd.DataFrame()  # Initialize an empty DataFrame if no data

    def load_basket(self):
        """Load basket from CSV"""
        if os.path.exists(self.basket_file):
            self.basket = pd.read_csv(self.basket_file)
        else:
            self.basket = pd.DataFrame(columns=["item", "quantity"])  # Create empty DataFrame

    def save_inventory_df(self, inventory_df, filename="updated_inventory.json"):
        """Save the updated inventory DataFrame to a new JSON file without modifying the original."""
        inventory_df.to_json(filename, orient="records", indent=4)
        print(f" Inventory saved to '{filename}' (original JSON remains unchanged).")

    def save_basket(self):
        """Save user basket back to CSV"""
        self.basket.to_csv(self.basket_file, index=False)

    def add_item(self, item, quantity):
        """Check if item exists, validate quantity of item left, and update both basket and inventory."""
        item = item.lower().strip()
        self.inventory_df["item_lower"] = self.inventory_df["item"].str.lower().str.strip()

        #  Step 1: Check if item exists in inventory
        if item not in self.inventory_df["item_lower"].values:
            print(f" '{item}' is not available in the store. Please choose a valid item.")
            return

        item_row = self.inventory_df[self.inventory_df["item_lower"] == item].iloc[0]
        available_quantity = item_row["quantity"]
        max_quantity = item_row["max_quantity"]

        # Step 2: Check if enough stock (quantity) is available
        if available_quantity == 0:
            print(f" '{item}' is out of stock!")
            return

        # Step 3: Enforce purchase limit
        if quantity > max_quantity:
            print(f" You can only buy up to {max_quantity} of '{item}' per order.")
            return

        # Step 4: Ensure stock is available for the requested quantity
        if quantity > available_quantity:
            print(f" Only {available_quantity} of '{item}' are left in stock.")
            return

        # Step 5: Add item to basket (or update quantity)
        if item in self.basket["item"].values:
            self.basket.loc[self.basket["item"] == item, "quantity"] += quantity
        else:
            new_entry = pd.DataFrame({"item": [item], "quantity": [quantity]})
            self.basket = pd.concat([self.basket, new_entry], ignore_index=True)

        # Step 6: Update inventory (reduce stock)
        self.inventory_df.loc[self.inventory_df["item_lower"] == item, "quantity"] -= quantity

        print(f" Added {quantity}x '{item}' to your basket.")
        self.save_inventory_df(self.inventory_df)
        self.save_basket()

    def remove_item(self, item):
        """Remove an item from the basket, restoring stock"""
        if item not in self.basket["item"].values:
            print(f" {item} is not in your basket.")
            return

        basket_quantity = self.basket.loc[self.basket["item"] == item, "quantity"].values[0]
        self.basket = self.basket[self.basket["item"] != item]  # Remove item
        self.inventory_df.loc[self.inventory_df["item_lower"] == item, "quantity"] += basket_quantity  # Restore full stock
        print(f" Removed all {basket_quantity} {item}(s) from basket.")

        self.save_inventory_df(self.inventory_df)
        self.save_basket()

    def update_quantity(self, item, quantity):
        """Update the quantity of an item inside basket"""
        if item not in self.basket["item"].values:
            print(f" {item} is not in your basket.")
            return

        # update the quantity of a specific item in the self.basket DataFrame
        self.basket.loc[self.basket["item"] == item, "quantity"] = quantity
        print(f" Updated {item} quantity to {quantity}.")

    def view_basket(self):
        """Display the user's basket"""
        print("\n Your Basket:")

        if self.basket.empty:
            print("Basket is empty.")
            print()
        else:
            print(self.basket)

    def clear_basket(self):
        """Clear all items in the basket"""
        self.basket = self.basket.iloc[0:0]  # Clear all rows, keep columns
        print(" Basket cleared.")

    def checkout(self):
        """Proceed to checkout, finalizing the basket"""
        print("\nProceeding to checkout...")
        print(f"Total Items in Basket: {len(self.basket)}")
        print(f"Final Basket: \n{self.basket}")

        # Logic for payment and finalizing checkout goes here

        # Generate receipt
        total_cost = generate_receipt(self.basket, self.inventory_df)

        # Log transaction
        log_transaction(self.basket, total_cost, self.inventory_df)

        # Clear basket after checkout
        self.clear_basket()

        # Display updated inventory
        print("\n=== Updated Inventory ===")
        display(self.inventory_df)

        # Save updated inventory to file
        self.save_inventory_df(self.inventory_df)
```

```python
# Main Loop
basket = GroceryBasket()  # initialize an instance of the GroceryBasket class

validOptionList = ['1', '2', '3', '4', '5', '6']
inputList = ["Add item", "Remove item", "Update quantity", "View basket", "Clear basket", "Checkout"]

while True:
    print("\n--- Choose an option ---")
    print("1 = Add item, 2 = Remove item, 3 = Update quantity, 4 = View basket, 5 = Clear basket, 6 = Checkout")

    try:
        num = input("Enter a number: ").strip()  # Get input and remove extra spaces

        if num not in validOptionList:  # Check if input is valid
            print("Invalid input! Please enter a number from 1 to 6.")
            continue  # Restart the loop

        print(f"You selected: {inputList[int(num) - 1]}")  # Show selected action

        if num == '1':  # User chose "Add item"
            item = input("Enter the item name: ").strip()
            item = item.lower() # lower case the input for what user wants to put in basket
            quantity = int(input("Enter the quantity: "))
            basket.add_item(item, quantity)

        elif num == '2':  # User chose "Remove item"
            item = input("Enter the item name to remove: ").strip()
            item = item.lower() # lower case the input for what user wants to put in basket
            basket.remove_item(item)

        elif num == '3':  # User chose "Update quantity"
            item = input("Enter the item name to update: ").strip()
            item = item.lower() # lower case the input for what user wants to put in basket

            quantity = int(input("Enter the new quantity: "))
            basket.update_quantity(item, quantity)

        elif num == '4':  # User chose "View basket"
            basket.view_basket()

        elif num == '5':  # User chose "Clear basket"
            basket.clear_basket()

        elif num == '6':  # User chose "Checkout"
            print("Exiting... Thank you for shopping! ")
            basket.checkout()
            break  # Exit the loop

    except ValueError:
        print("Invalid input! Please enter a valid number.")
    except Exception as e:  # Catch unexpected errors
        print(f"An error occurred: {e}. Please try again.")
```
